```python
# 2. Write a Python function that takes a list and returns a new list with distinct elements from the first list.

def distinct(l):
    new = []
    for i in l:
        if i not in new:
         new.append(i)
    return new

l=[1,2,3,3,4,2,5]
distinct_list = distinct(l)
print(distinct_list)
```

```python
# 4. Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.

def factorial(n):

    if n<0:
       return "enter positive integer"
    fact = 1
    for i in range(1,n+1):
     fact *= i
    return fact

print(factorial(7))
    

```

```python
# 6. Write a Python function to check whether a number falls within a given range.

def inrange(n,start,end):
    if(n>= start and n<=end):
     return "number is in range"
    
    else:
       return "number not in range"
    
print(inrange(4,5,6))
```

```python
# 7. Write a Python function to Print Even Numbers from a Given List

def even(l):
    for i in l:
        if(i%2 == 0):
            print(i)
    
l=[1,6,34,3,7,9]
even(l)
```

```python
# 8. Write a Python function that takes a number as a parameter and checks whether the number is prime or not.

def isPrime(n):
    if n<=1:
      return "non-prime"
    
    for i in range(2,n):
       if(n%i ==0):
          return "non-prime"
    return "prime"

print(isPrime(6))
print(isPrime(139))
```

```python
# 9. Write a Python function that accepts a string and counts the number of upper and lower case letters.

def counting(str):
   lowercase = 0
   uppercase = 0
   
   for i in str:
      if i.islower():
         lowercase += 1
      elif i.isupper():
         uppercase += 1

   print("lowercase=", lowercase)
   print("uppercase=",uppercase)
 
counting("eLl")

```
