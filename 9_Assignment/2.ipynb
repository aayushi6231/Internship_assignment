```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Name':['a1','a2','a3'],
    'Marks':[78,90,np.nan],
    'City':['Delhi','Mumbai','Delhi'],
    'Date':pd.to_datetime([
        '2026-05-21',
        '2026-05-22',
        '2026-05-23'
    ])
})

print(df)

print(df.isnull().sum())

x = df['Marks'].mean()

df.fillna({'Marks':x}, inplace=True)

print(df)

print(df['Marks'].mean())

print(df['Marks'].max())

print(df['Marks'].min())

gd = df.groupby('City')['Marks'].mean()
print(gd)

res = df.sort_values(by='Marks')
print(res)
```
