```python
import pandas as pd

df1 = pd.DataFrame({
    "empID": [101, 102],
    "name": ["A1", "A2"]
})

df2 = pd.DataFrame({
    "EmpID": [103, 104],
    "Name": ["A3", "A4"]
})

df3 = pd.DataFrame({
    "empID": [101, 102, 103, 104],
    "Department": ["cs", "IT", "Finance", "Marketing"]
})
```

```python
concat_df = pd.concat([df1, df2])
print(concat_df)
```

```python
merged_df = pd.merge(concat_df, df3, on="empID")

print("\nMerged:")
print(merged_df)
```
