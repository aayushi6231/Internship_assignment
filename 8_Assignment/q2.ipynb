```python
import pandas as pd
df1 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "name": ["a1", "a2", "a3", "a4"]
})

df2 = pd.DataFrame({
    "id": [2, 3, 4, 5],
    "city": ["delhi", "mumbai", "kota", "Jaipur"]
})
print(df1)
print(df2)
```

```python
inner_merge= pd.merge(df1, df2, on="id", how="inner")

print(inner_merge)
```

```python
left_join = pd.merge(df1, df2, on="id", how="left")

print(left_join)
```

```python
right_join = pd.merge(df1, df2, on="id", how="right")

print(right_join)
```

```python
df1_index = df1.set_index("id")
df2_index = df2.set_index("id")

index_join = df1_index.join(df2_index)

print(index_join)
```

```python
df3 = pd.DataFrame({
    "id": [1, 1, 2, 2],
    "Dept": ["cs", "IT", "cs", "IT"],
    "Salary": [30000, 40000, 35000, 45000]
})

df4 = pd.DataFrame({
    "id": [1, 2, 2],
    "Dept": ["IT", "cs", "IT"],
    "Bonus": [5000, 4000, 6000]
})

multi_merge = pd.merge(df3, df4, on=["id", "Dept"], how="inner")

print(multi_merge)
```
