```python
import csv
with open("address_book.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Address", "Mob", "Email"])

    for i in range(2):
        print(f"\nEnter details of person {i+1}")
        name = input("Name: ")
        address = input("Address: ")
        mob = input("Mob: ")
        email = input("Email: ")

        writer.writerow([name, address, mob, email])
```
